### Statistical Measures of the database

#### Maximum and Minimum in each column for analysis

```python
x = db['Total Library Size'].max()
y = db['Total Library Size'].idxmax()
z = db.loc[y,'Country']
print(z,":",x)
```

#### The above result shows that Czechia has the highest amount of Library size of 7325 and can stream large amount of shows / movies on Netflix

```python
x = db['Total Library Size'].min()
y = db['Total Library Size'].idxmin()
z = db.loc[y,'Country']
print(z,":",x)
```

#### The above result shows that Croatia has the least Library size of 2274 , this implies that they'll have to remove old movies / shows which people won't watch and constantly update their library.

### Series representation of the Database

```python
s = pd.Series(data=db['No. of TV Shows'].values,index=db['Country'])
print(s)
```

#### Graphical Representations

```python
s.plot(figsize=(20,30),y='Number of TV Shows')
```

#### Bar Graph for the Series

```python
s.plot.bar(figsize=(20,30),y="Number of TV Shows")
```

#### Pie Chart for the Series 

```python
s.plot.pie(figsize=(10,20))
```

#### Histogram for the Series

```python
s.plot.hist(figsize=(10,20))
```

#### Area plot for the Series

```python
s.plot.area(figsize=(20,30))
```
